```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier

# Load the iris dataset
iris = datasets.load_iris()
x = iris.data[:, :2]  # X-Axis - sepal length-width
y = iris.target       # Y-Axis - species

# Define the boundaries of the mesh
x_min, x_max = x[:, 0].min() - 0.5, x[:, 0].max() + 0.5
y_min, y_max = x[:, 1].min() - 0.5, x[:, 1].max() + 0.5

# Define the mesh grid step size and create the mesh grid
h = 0.02
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Initialize and fit the K-Nearest Neighbors classifier
knn = KNeighborsClassifier()
knn.fit(x, y)

# Predict the class for each point in the mesh grid
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Create the plot
plt.figure()
cmap_light = ListedColormap(['#AAAAFF', '#AAFFAA', '#FFAAAA'])
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

# Plot the training points
plt.scatter(x[:, 0], x[:, 1], c=y, edgecolor='k', marker='o')

# Set the limits of the plot
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model, datasets

# Load the diabetes dataset
diabetes = datasets.load_diabetes()

# Split the data into training and testing sets
x_train = diabetes.data[:-20]
y_train = diabetes.target[:-20]
x_test = diabetes.data[-20:]
y_test = diabetes.target[-20:]

# Extract the feature of interest and reshape for fitting
x0_test = x_test[:, 0][:, np.newaxis]
x0_train = x_train[:, 0][:, np.newaxis]

# Initialize and fit the Linear Regression model
linreg = linear_model.LinearRegression()
linreg.fit(x0_train, y_train)

# Make predictions
y_pred = linreg.predict(x0_test)

# Plot the results
plt.scatter(x0_test, y_test, color='k', label='Test data')
plt.plot(x0_test, y_pred, color='b', linewidth=3, label='Linear regression')
plt.xlabel('Feature value')
plt.ylabel('Target value')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model, datasets

# Load the diabetes dataset
diabetes = datasets.load_diabetes()

# Split the data into training and testing sets
x_train = diabetes.data[:-20]
y_train = diabetes.target[:-20]
x_test = diabetes.data[-20:]
y_test = diabetes.target[-20:]

# Initialize Linear Regression model
linreg = linear_model.LinearRegression()

# Set up the plot
plt.figure(figsize=(8, 12))

# Iterate over each feature
for f in range(10):
    xi_test = x_test[:, f][:, np.newaxis]
    xi_train = x_train[:, f][:, np.newaxis]
    
    # Fit the model and predict
    linreg.fit(xi_train, y_train)
    y_pred = linreg.predict(xi_test)
    
    # Plot results
    plt.subplot(5, 2, f + 1)
    plt.scatter(xi_test, y_test, color='k', label='Test data')
    plt.plot(xi_test, y_pred, color='b', linewidth=3, label='Linear regression')
    plt.xlabel('Feature value')
    plt.ylabel('Target value')
    plt.title(f'Feature {f + 1}')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets

# Load the diabetes dataset
diabetes = datasets.load_diabetes()

# Split the data into training and testing sets
x_train = diabetes.data[:-20]
y_train = diabetes.target[:-20]
x_test = diabetes.data[-20:]
y_test = diabetes.target[-20:]

# Extract and preprocess the feature of interest
x0_test = x_test[:, 2][:, np.newaxis] * 100
x0_train = x_train[:, 2][:, np.newaxis] * 100

# Sort x0_test for better plotting
x0_test.sort(axis=0)

# Initialize and fit the Support Vector Regression models
svr = svm.SVR(kernel='linear', C=1000)
svr2 = svm.SVR(kernel='poly', C=1000, degree=2)
svr3 = svm.SVR(kernel='poly', C=1000, degree=3)

svr.fit(x0_train, y_train)
svr2.fit(x0_train, y_train)
svr3.fit(x0_train, y_train)

# Make predictions
y = svr.predict(x0_test)
y2 = svr2.predict(x0_test)
y3 = svr3.predict(x0_test)

# Plot the results
plt.scatter(x0_test, y_test, color='k', label='Test data')
plt.plot(x0_test, y, color='b', label='Linear SVR')
plt.plot(x0_test, y2, color='r', label='Polynomial SVR (degree=2)')
plt.plot(x0_test, y3, color='g', label='Polynomial SVR (degree=3)')

# Add labels and legend
plt.xlabel('Feature value (scaled)')
plt.ylabel('Target value')
plt.title('Support Vector Regression')
plt.legend()

plt.show()
```
