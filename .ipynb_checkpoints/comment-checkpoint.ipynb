```python
import re
import requests
import pandas as pd

def get_cmt_from_shopee_url(url):
    r = re.search(r"i\.(\d+)\.(\d+)", url)
    if not r:
        return {"comment": [], "rating": []}

    shop_id, item_id = r[1], r[2]
    ratings_url = "https://shopee.vn/api/v2/item/get_ratings?filter=0&flag=1&itemid={item_id}&limit=50&offset={offset}&shopid={shop_id}&type={rating}"

    d = {"comment": [], "rating": []}
    
    # Add headers to mimic a request from a browser
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36"
    }
    for rate in range(1, 6):
        offset = 0
        while True:
            try:
                # Add headers to the GET request
                response = requests.get(ratings_url.format(shop_id=shop_id, item_id=item_id, offset=offset, rating=rate), headers=headers)
                response.raise_for_status()  # Raises an HTTPError for bad responses (4xx, 5xx)
                
                data = response.json()

                # Ensure there is data before processing
                ratings = data.get("data", {}).get("ratings", [])
                if not ratings:
                    break

                for rating in ratings:
                    d["rating"].append(rating["rating_star"])
                    d["comment"].append(rating["comment"])

                # If there are less than 50 ratings, we can stop
                if len(ratings) < 50:
                    break

                offset += 50  # Increase offset to get the next set of ratings

            except requests.exceptions.RequestException as e:
                print(f"Error fetching data: {e}")
                break  # Exit the loop in case of network or API error

    return d

# Example URLs
url1 = "https://shopee.vn/%C3%81o-s%C6%A1-mi-tay-ng%E1%BA%AFn-nam-n%E1%BB%AF-form-r%E1%BB%99ng-s%C6%A1-mi-c%E1%BB%95-vest-unisex-tay-l%E1%BB%A1-ch%E1%BA%A5t-v%E1%BA%A3i-l%E1%BB%A5a-m%E1%BB%8Bn-ch%C3%B4ng-nh%C4%83n-i.310912000.5852968977"
url2 = "https://shopee.vn/%C3%81o-s%C6%A1-mi-tay-d%C3%A0i-nam-n%E1%BB%AF-d%C3%A1ng-r%E1%BB%99ng-cao-c%E1%BA%A5p-v%E1%BA%A3i-nhung-t%C4%83m-DAVUBA-SM004-i.753272949.13190051003"

# Getting comments from URLs
comments_data = get_cmt_from_shopee_url(url1)  # Replace with the URL of your choice

# Convert the dictionary into a pandas DataFrame
df = pd.DataFrame(comments_data)

# Export the DataFrame to a CSV file
df.to_csv('comments_and_ratings.csv', index=False)

print("CSV file has been saved successfully.")
```

```python
import pandas as pd
import re
from sklearn.model_selection import train_test_split

# Load dữ liệu
data = pd.read_csv("comments_and_ratings.csv")  # Thay bằng đường dẫn file của bạn

# Bước 1: Loại bỏ dữ liệu không cần thiết
def clean_text(text):
    if not isinstance(text, str):
        return ""
    text = re.sub(r"\n", " ", text)  # Thay \n bằng khoảng trắng
    text = re.sub(r"[^\w\s]", "", text)  # Loại bỏ ký tự đặc biệt
    text = re.sub(r"\s+", " ", text).strip()  # Xóa khoảng trắng thừa
    return text.lower()  # Chuyển về chữ thường

data['comment'] = data['comment'].apply(clean_text)

# Bước 2: Xóa dữ liệu bị thiếu hoặc không hợp lệ
data = data.dropna(subset=['comment', 'rating'])  # Loại bỏ hàng bị thiếu

# Bước 3: Chuyển đổi rating về nhãn phân loại (1: tiêu cực, 3: trung lập, 5: tích cực)
def categorize_rating(rating):
    if rating <= 2:
        return "negative"
    elif rating == 3:
        return "neutral"
    else:
        return "positive"

data['label'] = data['rating'].apply(categorize_rating)

# Bước 4: Chia tập dữ liệu thành training và testing
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

# Lưu lại tập dữ liệu đã xử lý
train_data.to_csv("train_data.csv", index=False)
test_data.to_csv("test_data.csv", index=False)

print("Dữ liệu sau khi tiền xử lý:")
print(data.head())
```

```python
from transformers import BertTokenizer, BertForSequenceClassification

# Tải tokenizer và model BERT đã được huấn luyện sẵn
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=3)  # 3 nhãn: negative, neutral, positive
```

```python
import torch
from torch.utils.data import DataLoader, Dataset

class CommentDataset(Dataset):
    def __init__(self, comments, labels, tokenizer, max_len):
        self.comments = comments
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.comments)

    def __getitem__(self, idx):
        comment = self.comments[idx]
        label = self.labels[idx]

        # Tokenize và chuyển đổi thành tensor
        encoding = self.tokenizer(
            comment,
            truncation=True,
            padding="max_length",
            max_length=self.max_len,
            return_tensors="pt"
        )

        return {
            "input_ids": encoding["input_ids"].squeeze(0),
            "attention_mask": encoding["attention_mask"].squeeze(0),
            "label": torch.tensor(label, dtype=torch.long),
        }

# Chuẩn bị tập dữ liệu
def prepare_labels(label_column):
    label_mapping = {"negative": 0, "neutral": 1, "positive": 2}
    return label_column.map(label_mapping).values

train_dataset = CommentDataset(
    train_data["comment"].values,
    prepare_labels(train_data["label"]),
    tokenizer,
    max_len=128
)

test_dataset = CommentDataset(
    test_data["comment"].values,
    prepare_labels(test_data["label"]),
    tokenizer,
    max_len=128
)

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=16)
```

```python
from torch.utils.data import Dataset

class CommentDataset(Dataset):
    def __init__(self, comments, labels, tokenizer, max_len):
        self.comments = comments
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.comments)

    def __getitem__(self, index):
        comment = str(self.comments[index])  # Chuyển về kiểu chuỗi
        label = self.labels[index]
        
        # Tokenize
        encoding = self.tokenizer.encode_plus(
            comment,
            add_special_tokens=True,
            max_length=self.max_len,
            return_token_type_ids=False,
            padding="max_length",
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt'
        )
        
        return {
            'input_ids': encoding['input_ids'].squeeze(0),
            'attention_mask': encoding['attention_mask'].squeeze(0),
            'label': torch.tensor(label, dtype=torch.long)
        }
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, MaxPooling1D, LSTM, Dropout, Dense, Flatten

# Hyperparameters
embedding_dim = 128
max_length = 200  # Maximum sequence length
vocab_size = 10000  # Vocabulary size
filter_size = 128  # Number of filters for CNN
kernel_size = 3  # Size of the convolutional kernel
dropout_rate = 0.5
num_classes = 3  # Number of output classes (negative, neutral, positive)

# Define the model
model = Sequential([
    # Embedding layer
    Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_length),

    # 1st CNN Layer
    Conv1D(filters=filter_size, kernel_size=kernel_size, activation='relu'),
    MaxPooling1D(pool_size=2),

    # 2nd CNN Layer
    Conv1D(filters=filter_size, kernel_size=kernel_size, activation='relu'),
    MaxPooling1D(pool_size=2),

    # Dropout after CNN layers
    Dropout(dropout_rate),

    # LSTM Layer
    LSTM(70, return_sequences=True),
    Dropout(dropout_rate),

    # Additional LSTM Layer
    LSTM(70, return_sequences=False),
    Dropout(dropout_rate),

    # Dense Layers
    Dense(128, activation='relu'),
    Dropout(dropout_rate),

    Dense(64, activation='relu'),
    Dropout(dropout_rate),

    # Output layer
    Dense(num_classes, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',  # Use categorical_crossentropy for one-hot labels
    metrics=['accuracy']
)

# Model summary
model.summary()

# Placeholder: Assume X_train, y_train, X_test, y_test are preprocessed and ready
# model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Đọc dữ liệu
train_data['comment'] = train_data['comment'].fillna(" ")
test_data['comment'] = test_data['comment'].fillna(" ")

# Vector hóa dữ liệu
vectorizer = TfidfVectorizer(max_features=500)
X_train = vectorizer.fit_transform(train_data['comment'])
X_test = vectorizer.transform(test_data['comment'])

# Mã hóa nhãn
label_mapping = {"negative": 0, "neutral": 1, "positive": 2}
y_train = train_data['label'].map(label_mapping)
y_test = test_data['label'].map(label_mapping)

# Huấn luyện mô hình
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Tính MAE, MSE, RMSE
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

# Vẽ biểu đồ
metrics = {"MAE": mae, "MSE": mse, "RMSE": rmse}
plt.figure(figsize=(10, 6))
sns.barplot(x=list(metrics.keys()), y=list(metrics.values()), palette="coolwarm")
plt.title("So sánh các chỉ số MAE, MSE, RMSE", fontsize=16)
plt.ylabel("Giá trị", fontsize=14)
plt.xlabel("Chỉ số", fontsize=14)
plt.show()
```
