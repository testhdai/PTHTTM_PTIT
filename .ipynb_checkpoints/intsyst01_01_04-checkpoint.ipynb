```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Xử lý dữ liệu MNIST (Phân loại chữ số viết tay)
def mnist_model():
    # Tải dữ liệu MNIST
    (train_images, train_labels), (test_images, test_labels) = mnist.load_data()
    
    # Định dạng lại và chuẩn hóa dữ liệu (Giá trị pixel từ 0-255 được chia cho 255 để nằm trong khoảng 0-1)
    train_images = train_images.reshape((60000, 28, 28, 1)).astype('float32') / 255
    test_images = test_images.reshape((10000, 28, 28, 1)).astype('float32') / 255

    # Mã hóa one-hot cho nhãn (Để chuyển đổi nhãn sang dạng nhị phân, phù hợp với output của mạng nơ-ron)
    train_labels = to_categorical(train_labels)
    test_labels = to_categorical(test_labels)

    # Xây dựng mô hình CNN cho MNIST
    model = models.Sequential()
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))  # Lớp Conv2D đầu vào
    model.add(layers.MaxPooling2D((2, 2)))  # Lớp MaxPooling để giảm kích thước
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))  # Lớp Conv2D thứ 2
    model.add(layers.MaxPooling2D((2, 2)))  # MaxPooling lần 2
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))  # Lớp Conv2D thứ 3

    # Flatten và thêm các lớp Dense
    model.add(layers.Flatten())
    model.add(layers.Dense(64, activation='relu'))  # Lớp Dense ẩn
    model.add(layers.Dense(10, activation='softmax'))  # Lớp đầu ra với softmax để phân loại 10 chữ số

    # Biên dịch mô hình
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    # Huấn luyện mô hình
    model.fit(train_images, train_labels, epochs=5, batch_size=64, validation_split=0.2)

    # Đánh giá mô hình
    test_loss, test_acc = model.evaluate(test_images, test_labels)
    print(f"Độ chính xác của MNIST trên tập test: {test_acc}")

mnist_model()
```

```python
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras import layers, models

# Xử lý dữ liệu IMDB (Phân tích cảm xúc)
def imdb_model():
    # Tải dữ liệu IMDB (Dữ liệu đã được mã hóa bằng chỉ số từ vựng)
    (train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words=10000)

    # Đệm các chuỗi để đảm bảo độ dài đồng nhất (Pad sequences)
    maxlen = 256
    train_data = pad_sequences(train_data, maxlen=maxlen)
    test_data = pad_sequences(test_data, maxlen=maxlen)

    # Xây dựng mô hình RNN cho IMDB
    model = models.Sequential()
    model.add(layers.Embedding(10000, 128, input_length=maxlen))  # Lớp Embedding để chuyển từ ngữ thành vector
    model.add(layers.LSTM(128))  # Lớp LSTM để xử lý chuỗi
    model.add(layers.Dense(1, activation='sigmoid'))  # Lớp đầu ra dùng sigmoid cho phân loại nhị phân

    # Biên dịch mô hình
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    # Huấn luyện mô hình
    model.fit(train_data, train_labels, epochs=5, batch_size=128, validation_split=0.2)

    # Đánh giá mô hình
    test_loss, test_acc = model.evaluate(test_data, test_labels)
    print(f"Độ chính xác của IMDB trên tập test: {test_acc}")

imdb_model()
```

```python
from tensorflow.keras.datasets import boston_housing
from tensorflow.keras import models, layers

# Xử lý dữ liệu Boston Housing (Dự đoán giá nhà)
def boston_housing_model():
    # Tải dữ liệu Boston Housing
    (train_data, train_labels), (test_data, test_labels) = boston_housing.load_data()

    # Chuẩn hóa dữ liệu (Quan trọng cho bài toán hồi quy)
    mean = train_data.mean(axis=0)
    std = train_data.std(axis=0)
    train_data = (train_data - mean) / std
    test_data = (test_data - mean) / std

    # Xây dựng mô hình MLP (mạng neural truyền thẳng) cho dự đoán giá nhà
    model = models.Sequential()
    model.add(layers.Dense(64, activation='relu', input_shape=(train_data.shape[1],)))  # Lớp Dense đầu vào
    model.add(layers.Dense(64, activation='relu'))  # Lớp Dense thứ hai
    model.add(layers.Dense(1))  # Lớp đầu ra (Dự đoán giá nhà)

    # Biên dịch mô hình
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])  # Dùng MSE cho bài toán hồi quy

    # Huấn luyện mô hình
    model.fit(train_data, train_labels, epochs=10, batch_size=64, validation_split=0.2)

    # Đánh giá mô hình
    test_loss, test_mae = model.evaluate(test_data, test_labels)
    print(f"Độ lỗi MAE của Boston Housing trên tập test: {test_mae}")

boston_housing_model()
```

```python
# Import các thư viện cần thiết
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Xây dựng mô hình cho bộ dữ liệu MNIST
def mnist_model():
    # Tải dữ liệu MNIST
    (train_images, train_labels), (test_images, test_labels) = mnist.load_data()

    # Chuyển đổi hình dạng và chuẩn hóa dữ liệu (Giá trị pixel từ 0-255 được chia cho 255 để đưa về khoảng 0-1)
    train_images = train_images.reshape((60000, 28, 28, 1)).astype('float32') / 255
    test_images = test_images.reshape((10000, 28, 28, 1)).astype('float32') / 255

    # Chuyển nhãn sang dạng one-hot encoding (để sử dụng cho bài toán phân loại)
    train_labels = to_categorical(train_labels)
    test_labels = to_categorical(test_labels)

    # Xây dựng mô hình CNN cho phân loại chữ số viết tay
    model = models.Sequential()
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))  # Lớp tích chập (Conv2D)
    model.add(layers.MaxPooling2D((2, 2)))  # Lớp gộp tối đa (MaxPooling)
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))

    # Làm phẳng dữ liệu và thêm các lớp fully connected
    model.add(layers.Flatten())
    model.add(layers.Dense(64, activation='relu'))  # Lớp fully connected ẩn
    model.add(layers.Dense(10, activation='softmax'))  # Lớp đầu ra với 10 lớp (ứng với 10 chữ số)

    # Biên dịch mô hình với Adam optimizer và hàm mất mát categorical_crossentropy
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    # Huấn luyện mô hình với dữ liệu huấn luyện
    model.fit(train_images, train_labels, epochs=5, batch_size=64, validation_split=0.2)

    # Đánh giá mô hình với tập dữ liệu kiểm tra
    test_loss, test_acc = model.evaluate(test_images, test_labels)
    print(f"Độ chính xác của MNIST trên tập test: {test_acc}")

# Chạy mô hình MNIST
mnist_model()
```

```python
# Import các thư viện cần thiết
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Xây dựng mô hình cho bộ dữ liệu MNIST
def mnist_model():
    print("Đang tải dữ liệu MNIST...")
    # Tải dữ liệu MNIST
    (train_images, train_labels), (test_images, test_labels) = mnist.load_data()

    # Chuyển đổi hình dạng và chuẩn hóa dữ liệu (Chuyển đổi các giá trị pixel từ 0-255 về khoảng 0-1)
    train_images = train_images.reshape((60000, 28, 28, 1)).astype('float32') / 255
    test_images = test_images.reshape((10000, 28, 28, 1)).astype('float32') / 255

    # Chuyển nhãn sang dạng one-hot encoding (để sử dụng cho bài toán phân loại)
    train_labels = to_categorical(train_labels)
    test_labels = to_categorical(test_labels)

    # Xây dựng mô hình CNN cho MNIST
    model = models.Sequential()
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))  # Lớp tích chập đầu tiên
    model.add(layers.MaxPooling2D((2, 2)))  # Lớp gộp tối đa
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))  # Lớp tích chập thứ hai
    model.add(layers.MaxPooling2D((2, 2)))  # Lớp gộp tối đa thứ hai
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))  # Lớp tích chập thứ ba

    # Làm phẳng dữ liệu và thêm lớp fully connected
    model.add(layers.Flatten())
    model.add(layers.Dense(64, activation='relu'))  # Lớp fully connected ẩn
    model.add(layers.Dense(10, activation='softmax'))  # Lớp đầu ra với 10 lớp (ứng với 10 chữ số)

    # Biên dịch mô hình với Adam optimizer và hàm mất mát categorical_crossentropy
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    # Huấn luyện mô hình
    print("Đang huấn luyện mô hình MNIST...")
    model.fit(train_images, train_labels, epochs=5, batch_size=64, validation_split=0.2)

    # Đánh giá mô hình
    test_loss, test_acc = model.evaluate(test_images, test_labels)
    print(f"Độ chính xác của MNIST trên tập test: {test_acc:.4f}")

# Chạy mô hình MNIST
mnist_model()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns  # Thư viện seaborn để vẽ biểu đồ nhiệt

# Giả lập dữ liệu cho 10 epochs
epochs = np.arange(1, 11)  # 10 epochs
loss = np.random.rand(10) * 0.5  # Loss values
accuracy = np.random.rand(10) * 100  # Accuracy values
mae = np.random.rand(10) * 10  # Mean Absolute Error
mse = np.random.rand(10) * 100  # Mean Squared Error
rmse = np.sqrt(mse)  # Root Mean Squared Error

# Tạo 7 biểu đồ riêng biệt
plt.figure(figsize=(18, 12))

# Biểu đồ 1: Loss (Đường)
plt.subplot(4, 2, 1)
plt.plot(epochs, loss, label='Loss', marker='o', color='red')
plt.title('Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss Value')
plt.xticks(epochs)
plt.legend()
plt.grid()

# Biểu đồ 2: Accuracy (Đường)
plt.subplot(4, 2, 2)
plt.plot(epochs, accuracy, label='Accuracy', marker='o', color='purple')
plt.title('Accuracy over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
plt.xticks(epochs)
plt.legend()
plt.grid()

# Biểu đồ 3: MAE (Đường)
plt.subplot(4, 2, 3)
plt.plot(epochs, mae, label='MAE', marker='o', color='blue')
plt.title('Mean Absolute Error (MAE) over Epochs')
plt.xlabel('Epochs')
plt.ylabel('MAE Value')
plt.xticks(epochs)
plt.legend()
plt.grid()

# Biểu đồ 4: MSE (Đường)
plt.subplot(4, 2, 4)
plt.plot(epochs, mse, label='MSE', marker='o', color='green')
plt.title('Mean Squared Error (MSE) over Epochs')
plt.xlabel('Epochs')
plt.ylabel('MSE Value')
plt.xticks(epochs)
plt.legend()
plt.grid()

# Biểu đồ 5: RMSE (Đường)
plt.subplot(4, 2, 5)
plt.plot(epochs, rmse, label='RMSE', marker='o', color='orange')
plt.title('Root Mean Squared Error (RMSE) over Epochs')
plt.xlabel('Epochs')
plt.ylabel('RMSE Value')
plt.xticks(epochs)
plt.legend()
plt.grid()

# Biểu đồ 6: Biểu đồ cột cho Loss và Accuracy
plt.subplot(4, 2, 6)
bar_width = 0.35
index = np.arange(len(epochs))
plt.bar(index, accuracy, bar_width, label='Accuracy', color='purple')
plt.bar(index + bar_width, loss, bar_width, label='Loss', color='red')
plt.title('Accuracy vs Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Values')
plt.xticks(index + bar_width / 2, epochs)
plt.legend()
plt.grid()

# Biểu đồ 7: Biểu đồ nhiệt cho tất cả các giá trị
plt.subplot(4, 2, 7)
heatmap_data = np.array([loss, accuracy, mae, mse, rmse])
sns.heatmap(heatmap_data, annot=True, fmt=".2f", cmap="coolwarm", xticklabels=epochs, yticklabels=['Loss', 'Accuracy', 'MAE', 'MSE', 'RMSE'])
plt.title('Biểu đồ nhiệt của Loss, Accuracy, MAE, MSE, RMSE')

# Căn chỉnh và hiển thị biểu đồ
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns  # Dùng để vẽ biểu đồ nhiệt (heatmap)

# Giả lập dữ liệu cho 10 epochs
epochs = np.arange(1, 11)  # 10 epochs
loss = np.random.rand(10) * 0.5  # Loss values
accuracy = np.random.rand(10) * 100  # Accuracy values
mae = np.random.rand(10) * 10  # Mean Absolute Error
mse = np.random.rand(10) * 100  # Mean Squared Error
rmse = np.sqrt(mse)  # Root Mean Squared Error

# Tạo 5 biểu đồ riêng biệt
plt.figure(figsize=(14, 10))

# Biểu đồ 1: Loss
plt.subplot(3, 2, 1)
plt.plot(epochs, loss, label='Loss', marker='o', color='red')
plt.title('Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss Value')
plt.xticks(epochs)
plt.legend()
plt.grid()

# Biểu đồ 2: Accuracy
plt.subplot(3, 2, 2)
plt.plot(epochs, accuracy, label='Accuracy', marker='o', color='purple')
plt.title('Accuracy over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
plt.xticks(epochs)
plt.legend()
plt.grid()

# Biểu đồ 3: MAE
plt.subplot(3, 2, 3)
plt.plot(epochs, mae, label='MAE', marker='o', color='blue')
plt.title('Mean Absolute Error (MAE) over Epochs')
plt.xlabel('Epochs')
plt.ylabel('MAE Value')
plt.xticks(epochs)
plt.legend()
plt.grid()

# Biểu đồ 4: MSE
plt.subplot(3, 2, 4)
plt.plot(epochs, mse, label='MSE', marker='o', color='green')
plt.title('Mean Squared Error (MSE) over Epochs')
plt.xlabel('Epochs')
plt.ylabel('MSE Value')
plt.xticks(epochs)
plt.legend()
plt.grid()

# Biểu đồ 5: RMSE
plt.subplot(3, 2, 5)
plt.plot(epochs, rmse, label='RMSE', marker='o', color='orange')
plt.title('Root Mean Squared Error (RMSE) over Epochs')
plt.xlabel('Epochs')
plt.ylabel('RMSE Value')
plt.xticks(epochs)
plt.legend()
plt.grid()

# Biểu đồ cột để so sánh tất cả các giá trị trong cùng một biểu đồ
plt.figure(figsize=(10, 6))
width = 0.15  # Độ rộng của các cột
plt.bar(epochs - 2*width, loss, width=width, label='Loss', color='red')
plt.bar(epochs - width, accuracy / 100, width=width, label='Accuracy', color='purple')  # Chia tỉ lệ cho accuracy
plt.bar(epochs, mae, width=width, label='MAE', color='blue')
plt.bar(epochs + width, mse / 100, width=width, label='MSE', color='green')  # Chia tỉ lệ cho MSE
plt.bar(epochs + 2*width, rmse, width=width, label='RMSE', color='orange')

plt.title('Comparison of Metrics Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Value')
plt.legend()
plt.grid()
plt.show()

# Biểu đồ nhiệt (heatmap) để so sánh tất cả các giá trị
metrics = np.array([loss, accuracy, mae, mse, rmse])

# Chuẩn bị dữ liệu heatmap: mỗi hàng là một metric
metrics_normalized = metrics / metrics.max(axis=1, keepdims=True)  # Chuẩn hóa dữ liệu để so sánh tốt hơn

plt.figure(figsize=(10, 6))
sns.heatmap(metrics_normalized, annot=True, cmap='coolwarm', yticklabels=['Loss', 'Accuracy', 'MAE', 'MSE', 'RMSE'])
plt.title('Heatmap of Metrics Comparison')
plt.xlabel('Epochs')
plt.ylabel('Metrics')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Giả lập dữ liệu cho 10 epochs
epochs = np.arange(1, 11)  # 10 epochs
loss = np.random.rand(10) * 0.5  # Loss values
accuracy = np.random.rand(10) * 100  # Accuracy values
mae = np.random.rand(10) * 10  # Mean Absolute Error
mse = np.random.rand(10) * 100  # Mean Squared Error
rmse = np.sqrt(mse)  # Root Mean Squared Error

# Gộp tất cả các giá trị thành một mảng 2D
metrics = np.vstack([loss, accuracy, mae, mse, rmse])

# Tạo biểu đồ kết hợp đường, cột và biểu đồ nhiệt

# Tạo biểu đồ đường cho các giá trị
plt.figure(figsize=(14, 10))

plt.plot(epochs, loss, label='Loss', marker='o', color='red', linewidth=2)
plt.plot(epochs, accuracy / 100, label='Accuracy', marker='o', color='purple', linewidth=2)  # Chia tỉ lệ cho accuracy
plt.plot(epochs, mae, label='MAE', marker='o', color='blue', linewidth=2)
plt.plot(epochs, mse / 100, label='MSE', marker='o', color='green', linewidth=2)  # Chia tỉ lệ cho MSE
plt.plot(epochs, rmse, label='RMSE', marker='o', color='orange', linewidth=2)

plt.title('Comparison of Metrics Over Epochs (Line Chart)')
plt.xlabel('Epochs')
plt.ylabel('Values')
plt.xticks(epochs)
plt.legend()
plt.grid()
plt.show()

# Biểu đồ cột để so sánh tất cả các giá trị
plt.figure(figsize=(14, 8))
width = 0.15  # Độ rộng của các cột
plt.bar(epochs - 2*width, loss, width=width, label='Loss', color='red')
plt.bar(epochs - width, accuracy / 100, width=width, label='Accuracy', color='purple')  # Chia tỉ lệ cho accuracy
plt.bar(epochs, mae, width=width, label='MAE', color='blue')
plt.bar(epochs + width, mse / 100, width=width, label='MSE', color='green')  # Chia tỉ lệ cho MSE
plt.bar(epochs + 2*width, rmse, width=width, label='RMSE', color='orange')

plt.title('Comparison of Metrics Over Epochs (Bar Chart)')
plt.xlabel('Epochs')
plt.ylabel('Values')
plt.legend()
plt.grid()
plt.show()

# Biểu đồ nhiệt (heatmap) để so sánh tất cả các giá trị
metrics_normalized = metrics / metrics.max(axis=1, keepdims=True)  # Chuẩn hóa dữ liệu để so sánh tốt hơn

plt.figure(figsize=(10, 6))
sns.heatmap(metrics_normalized, annot=True, cmap='coolwarm', yticklabels=['Loss', 'Accuracy', 'MAE', 'MSE', 'RMSE'])
plt.title('Heatmap of Metrics Comparison')
plt.xlabel('Epochs')
plt.ylabel('Metrics')
plt.show()
```
