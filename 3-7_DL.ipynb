```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Bước 1: Tạo dữ liệu ngẫu nhiên cho 1000 sinh viên
np.random.seed(0)

num_students = 1000
age = np.random.randint(18, 30, size=num_students)  # Tuổi từ 18 đến 30
hours_of_study = np.random.randint(1, 10, size=num_students)  # Giờ học mỗi ngày
average_scores = np.random.randint(40, 100, size=num_students)  # Điểm trung bình trước đó

# Kết quả đầu ra: Pass = 1, Fail = 0
pass_fail = np.where((hours_of_study * average_scores / age) > 20, 1, 0)

# Gom các thuộc tính lại thành một ma trận đầu vào
X = np.column_stack ( (age, hours_of_study, average_scores) )
y = pass_fail

# Bước 2: Xây dựng mạng nơ-ron bằng TensorFlow thuần
# Đầu vào có 3 thuộc tính (age, hours_of_study, average_scores)
n_features = 3
n_classes = 2  # 2 lớp: Pass và Fail

class SimpleNeuralNetwork(tf.Module):
    def __init__(self):
        # Khởi tạo các trọng số và bias cho từng lớp
        self.w1 = tf.Variable(tf.random.normal([n_features, 8]), dtype=tf.float32)
        self.b1 = tf.Variable(tf.zeros([8]), dtype=tf.float32)
        self.w2 = tf.Variable(tf.random.normal([8, 8]), dtype=tf.float32)
        self.b2 = tf.Variable(tf.zeros([8]), dtype=tf.float32)
        self.w3 = tf.Variable(tf.random.normal([8, 8]), dtype=tf.float32)
        self.b3 = tf.Variable(tf.zeros([8]), dtype=tf.float32)
        self.w_out = tf.Variable(tf.random.normal([8, n_classes]), dtype=tf.float32)
        self.b_out = tf.Variable(tf.zeros([n_classes]), dtype=tf.float32)
    
    @tf.function
    def __call__(self, x):
        # Lớp ẩn 1
        z1 = tf.matmul(x, self.w1) + self.b1
        a1 = tf.nn.relu(z1)
        # Lớp ẩn 2
        z2 = tf.matmul(a1, self.w2) + self.b2
        a2 = tf.nn.relu(z2)
        # Lớp ẩn 3
        z3 = tf.matmul(a2, self.w3) + self.b3
        a3 = tf.nn.relu(z3)
        # Lớp đầu ra
        output = tf.matmul(a3, self.w_out) + self.b_out
        return output

# Bước 3: Hàm mất mát và tối ưu hóa
def compute_loss(logits, labels):
    return tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=tf.one_hot(labels, depth=n_classes), logits=logits))

optimizer = tf.optimizers.Adam(learning_rate=0.001)

# Bước 4: Vòng lặp huấn luyện
def train_step(model, inputs, labels):
    with tf.GradientTape() as tape:
        logits = model(inputs)
        loss = compute_loss(logits, labels)
    gradients = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))
    return loss

# Bước 5: Đánh giá mô hình
def compute_accuracy(logits, labels):
    predictions = tf.argmax(logits, axis=1, output_type=tf.int32)
    return tf.reduce_mean(tf.cast(tf.equal(predictions, labels), tf.float32))

# Bước 6: Huấn luyện mô hình
n_epochs = 100
batch_size = 32

# Chuyển đổi dữ liệu đầu vào sang dạng tensor
X = tf.convert_to_tensor(X, dtype=tf.float32)
y = tf.convert_to_tensor(y, dtype=tf.int32)

# Khởi tạo mô hình
model = SimpleNeuralNetwork()

# Vòng lặp huấn luyện
losses = []
accuracies = []

for epoch in range(n_epochs):
    # Chia batch và huấn luyện từng batch
    for i in range(0, num_students, batch_size):
        batch_x = X[i:i+batch_size]
        batch_y = y[i:i+batch_size]
        loss = train_step(model, batch_x, batch_y)
    
    # Đánh giá độ chính xác sau mỗi epoch
    logits = model(X)
    accuracy = compute_accuracy(logits, y)
    accuracies.append(accuracy.numpy())
    losses.append(loss.numpy())
    
    if epoch % 10 == 0:
        print(f'Epoch {epoch}: Loss = {loss.numpy()}, Accuracy = {accuracy.numpy()}')

# Bước 7: Vẽ biểu đồ đánh giá độ chính xác
plt.plot(range(n_epochs), accuracies)
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Model Accuracy over Epochs')
plt.show()
```
